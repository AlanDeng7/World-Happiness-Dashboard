```python
#Making the clean destination CSV to final usable format

# group by 'Destination' and compute the sum of all columns
destination_final = clean_data_csv.groupby('Destination').agg({'2015 Combined': 'sum', '2020 Combined': 'sum','2015 Males': 'sum', '2020 Males': 'sum','2015 Females': 'sum', '2020 Females': 'sum'})

# convert the columns to rows
destination_final = destination_final.stack().reset_index()

# rename the columns
destination_final.columns = ['Destination', 'Year', 'Migrating into the country']

# print the resulting dataframe
destination_final
```
